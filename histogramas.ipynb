```python
import pandas as pd
```

```python
ruta = "C:\\Users\\WIN10\\Desktop\\MAESTRO EN VISUALIZACION DE DATOS\\03 - titanic.csv"
```

```python
df = pd.read_csv(ruta)
df = df.dropna(subset=["Age"], axis = "index") ## eliminar valores faltantes de "Age"
df.head(30)
```

```python
df.hist(
    "Age"

)
```

```python
df.hist(
    "Age",
    by = "Sex",
    bins = 20

)
```

```python
import seaborn as sns
```

```python
sns.histplot(
    df,
    x = "Age",
    kde = True,
    bins = 30   ### el bins se puede colocar desde cero hacia adelante
    

)
```

```python
sns.histplot(
    df,
    x = "Age",
    kde = True,
    bins = 30,   ### el bins se puede colocar desde cero hacia adelante
    hue = "Sex"

)
```

```python
sns.histplot(
    df,
    x = "Age",
    cumulative = True,

    bins = 20,  ### el bins se puede colocar desde cero hacia adelante
    fill = False

)
```

```python
sns.histplot(
    df,
    x = "Age",
    cumulative = True,
    element = "step",
    bins = 20,  ### el bins se puede colocar desde cero hacia adelante
    fill = False

)
```

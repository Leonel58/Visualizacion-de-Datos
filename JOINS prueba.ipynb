```python
DATA_1 = pd.read_csv(R_1)
DATA_2=pd.read_csv(R_2)
DATA_3=pd.read_csv(R_3)
```

```python
DATA_3.merge(DATA_2, on = "user", how = "left")
```

```python
DATA_C_USUARIOS.merge(DATA_1, on = "ciudad", how = "left")
```

```python
DATA_COMPLETADO = DATA_C_USUARIOS.merge(DATA_1, on = "ciudad", how = "left")
```

```python
DATA_COMPLETADO.columns
```

```python
DATA_COMPLETADO
```

```python
DATA_COMPLETADO["fecha"] = pd.to_datetime(DATA_COMPLETADO["fecha_publicado"], format = "%d/%m/%Y" )
DATA_COMPLETADO.head(5)
```

```python
DATA_COMPLETADO.columns
```

```python
DATA_COMPLETADO["Dia_Semana"] = DATA_COMPLETADO["fecha"].apply(lambda x: x.strftime("%A"))
DATA_COMPLETADO.head(10)
```

```python
DATA_COMPLETADO["mes"] = DATA_COMPLETADO["fecha_publicado"].apply(lambda x: x[3:])
DATA_COMPLETADO.head(5)
```

```python
DATA_COMPLETADO.groupby("mes").agg({
    
    "id":"count"
})
```

```python
AGRUPADO = DATA_COMPLETADO.groupby(["mes","ciudad"]).agg({
    "id":"count"
    
    
})

AGRUPADO
```

```python
AGRUPADO.reset_index("ciudad")

```

```python
DESAGRUPADO = AGRUPADO.reset_index("ciudad")
DESAGRUPADO
```

```python
DESAGRUPADO[["ciudad"]]
```

```python
DESAGRUPADO.pivot(columns = "ciudad")
```

```python
PIVOTE = DESAGRUPADO.pivot(columns = "ciudad")
```
